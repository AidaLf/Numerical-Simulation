## Problem 7: Plot the ratio of un-ionized donor atoms to the total electron concentration versus temperature for silicon over the range 50k<T<200k.


```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
Nc=2.8*1000
Nd=1
ionizedE=-0.045
def ndRatio(T):
  a=Nc*math.sqrt((T/300)**3)/(2*Nd)
  b=math.exp(ionizedE/(0.0259*(T/300)))
  return 1/(1+a*b)


```

```python
temp=np.arange(50,200,1)
ratio=[]
for t in temp:
  r=ndRatio(t)
  ratio.append(r)
plt.plot(temp,ratio)
plt.title("The ratio of un-ionized donor atoms to the total electron concentration versus temperature for silicon")
plt.xlabel("Temperature (K)")
plt.ylabel(r'$\frac{n{d}}{n{d}+n{0}}$')
plt.show()
```
